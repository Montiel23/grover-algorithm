```python
def multi_qubit_sum(qc, multi_qubs, ancillas, pre_constraint_len):
    for idx, comb in enumerate(multi_qubs):
        ancilla_idx = pre_constraint_len + idx

        qc.cx(comb[0], ancillas[ancilla_idx])
        qc.cx(comb[1], ancillas[ancilla_idx])
        qc.cx(comb[2], ancillas[ancilla_idx])

        qc.ccx(comb[0], comb[1], ancillas[ancilla_idx + 1])
        qc.ccx(comb[2], ancillas[ancilla_idx + 1], ancillas[ancilla_idx])

        #freeing up qubits to 0
        qc.ccx(comb[0], comb[1], ancillas[ancilla_idx + 1])

        # qc.x(ancillas[ancilla_idx + 1])
        # qc.ccx(comb[2], ancillas[ancilla_idx + 1], ancillas[ancilla_idx])

        qc.barrier()
        
```

```python
def three_qubit_diagonal_constraints_opt(qc, qubits, qub_diag, ancillas, pre_constraint_len):
    temp_ancilla = ancillas[pre_constraint_len]

    for idx, comb in enumerate(qub_diag):
        if idx == 1:
            ancilla_idx += 2

        else:
            ancilla_idx = idx + pre_constraint_len

        qc.cx(comb[0], ancillas[ancilla_idx])
        qc.cx(comb[1], ancillas[ancilla_idx])
        qc.cx(comb[2], ancillas[ancilla_idx])

        qc.ccx(comb[0], comb[1], ancillas[ancilla_idx + 1])
        qc.ccx(comb[2], ancillas[ancilla_idx + 1], ancillas[ancilla_idx + 2])

        qc.cx(ancillas[ancilla_idx + 2], ancillas[ancilla_idx])

        qc.ccx(comb[2], ancillas[ancilla_idx + 1], ancillas[ancilla_idx + 2])
        qc.ccx(comb[0], comb[1], ancillas[ancilla_idx + 1])

        qc.x(comb[0])
        qc.x(comb[1])
        qc.x(comb[2])

        qc.ccx(comb[0], comb[1], ancillas[ancilla_idx + 1])
        qc.ccx(comb[2], ancillas[ancilla_idx + 1], ancillas[ancilla_idx + 2])

        qc.cx(ancillas[ancilla_idx + 2], ancillas[ancilla_idx])

        qc.ccx(comb[2], ancillas[ancilla_idx + 1], ancillas[ancilla_idx + 2])
        qc.ccx(comb[0], comb[1], ancillas[ancilla_idx + 1])

        qc.barrier()
```

```python
def four_qubit_diagonal_constraints(qc, qub_diag, ancillas, pre_constraint_len):
    temp_ancilla = ancillas[pre_constraint_len]
    for idx, comb in enumerate(qub_diag):
        qc.cx(comb[0], temp_ancilla)
        qc.cx(comb[1], temp_ancilla)
        qc.cx(comb[2], temp_ancilla)
        qc.cx(comb[3], temp_ancilla)

        qc.ccx(comb[0], comb[1], ancillas[pre_constraint_len + 1])
        qc.ccx(comb[2], comb[3], ancillas[pre_constraint_len + 2])
        qc.ccx(ancillas[pre_constraint_len + 1], ancillas[pre_constraint_len + 2], temp_ancilla)

        qc.mcx([comb[0], comb[1], comb[2]], temp_ancilla)
        qc.mcx([comb[0], comb[1], comb[3]], temp_ancilla)
        qc.mcx([comb[0], comb[2], comb[3]], temp_ancilla)
        qc.mcx([comb[1], comb[2], comb[3]], temp_ancilla)

        qc.mcx([comb[0], comb[1], comb[2], comb[3]], ancillas[pre_constraint_len + 3])
        qc.cx(ancillas[pre_constraint_len + 3], temp_ancilla)

        qc.x(comb[0])
        qc.x(comb[1])
        qc.x(comb[2])
        qc.x(comb[3])

        qc.mcx([comb[0], comb[1], comb[2], comb[3]], ancillas[pre_constraint_len + 4])
        qc.cx(ancillas[pre_constraint_len + 4], temp_ancilla)

        #freeing up qubits to 0
        
        
        qc.barrier()
```

```python
def check_all_one(qc, ancil_check, ancillas, pre_constraint_len):

    qc.cx(ancil_check[0], ancillas[pre_constraint_len])

    for idx in range(1, len(ancil_check)):
        # qc.cx(ancillas[idx], pre_constraint_len + idx)
        qc.ccx(ancil_check[idx], ancillas[pre_constraint_len + idx - 1], ancillas[pre_constraint_len + idx])

    qc.barrier()

    # qc.cx(pre_constraint_len + len(ancillas), pre_constraint_len + len(ancillas) + 1)
    qc.cx(ancillas[pre_constraint_len + len(ancil_check) -1], ancillas[pre_constraint_len + len(ancil_check)])

    qc.barrier()

    for idx in reversed(range(1, len(ancil_check))):
        qc.ccx(ancil_check[idx], ancillas[pre_constraint_len + idx - 1], ancillas[pre_constraint_len + idx])

    qc.cx(ancil_check[0], ancillas[pre_constraint_len])

    qc.barrier()
```

```python
def check_all_one_efficient(qc, constraint_qubits, ancilla_a, ancilla_b, phase_flip_target=None, uncompute=False):
    if len(constraint_qubits) < 2:
        raise ValueError('Need at least 2 constraint qubits')

    if not uncompute:
        qc.ccx(constraint_qubits[0], constraint_qubits[1], ancilla_a)

        for idx in range(2, len(constraint_qubits)):
            qc.ccx(constraint_qubits[idx], ancilla_a, ancilla_b)
            qc.cx(ancilla_b, ancilla_a)
            qc.ccx(constraint_qubits[idx], ancilla_a, ancilla_b)

        if phase_flip_target is not None:
            inst = qc.z(phase_flip_target)
            inst.c_if(qc.cregs[0], 1)

    else:
        for idx in reversed(range(2, len(constraint_qubits))):
            qc.ccx(constraint_qubits[idx], ancilla_a, ancilla_b)
            qc.cx(ancilla_b, ancilla_a)
            qc.ccx(constraint_qubits[idx], ancilla_a, ancilla_b)

        qc.ccx(constraint_qubits[0], constraint_qubits[1], ancilla_a)

    qc.barrier()
```

```python
def grover_diff(qc, qubits, ancillas, pre_constraint_len):
    n = len(qubits)
    anc_start = pre_constraint_len
    anc_idx = anc_start
    print(anc_idx)

    qc.h(qubits)
    qc.x(qubits)
    qc.barrier()

    qc.ccx(qubits[0], qubits[1], ancillas[anc_idx])
    anc_idx += 1

    for i in range(2, n-1):
        qc.ccx(qubits[i], ancillas[anc_idx - 1], ancillas[anc_idx])
        anc_idx += 1

    qc.barrier()
    qc.cz(qubits[-1], ancillas[anc_idx - 1])
    qc.barrier()

    for i in reversed(range(2, n-1)):
        anc_idx -= 1
        qc.ccx(qubits[i], ancillas[anc_idx -1], ancillas[anc_idx])

    anc_idx -= 1
    qc.ccx(qubits[0], qubits[1], ancillas[anc_idx])
    qc.barrier()

    qc.x(qubits)
    qc.h(qubits)
    qc.barrier()
```

```python
def grover_diff_efficient(qc, qubits):
    qc.h(qubits)
    qc.x(qubits)

    qc.barrier()

    qc.h(qubits[-1])
    qc.mcx(qubits[:-1], qubits[-1])
    qc.h(qubits[-1])
    
    qc.barrier()

    qc.x(qubits)
    qc.h(qubits)
    qc.barrier()
```
